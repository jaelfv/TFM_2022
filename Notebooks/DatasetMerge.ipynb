# Dataset merge
In this notebook we are going to create the different files used for the models. This means we will generate the files after merging with some extermnal datasets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

We import the files to be merged

```python
data = pd.read_csv('../Data/TFM_Hdataset_clean.csv', index_col=0)
raw_boots = pd.read_csv("../Data/GT_boots.csv",skiprows=1,infer_datetime_format=True)
raw_sandals = pd.read_csv("../Data/GT_sandals.csv",skiprows=1,infer_datetime_format=True)
raw_goretex = pd.read_csv("../Data/GT_goretex.csv",skiprows=1,infer_datetime_format=True)
raw_slippers = pd.read_csv("../Data/GT_slippers.csv",skiprows=1,infer_datetime_format=True)
raw_formalshoes = pd.read_csv("../Data/GT_formalshoes.csv",skiprows=1,infer_datetime_format=True)
```

```python
raw_boots['style']='Boots'
raw_sandals['style']='Sandals'
raw_goretex['style']='Goretex'
raw_slippers['style']='Slippers'

raw_boots=raw_boots.rename(columns={'Setmana': 'Date','boots: (Regne Unit)':'boots'})
raw_sandals=raw_sandals.rename(columns={'Setmana': 'Date','sandals: (Regne Unit)':'sandals'})
raw_goretex=raw_goretex.rename(columns={'Setmana': 'Date','Goretex: (Regne Unit)':'goretex'})
raw_slippers=raw_slippers.rename(columns={'Setmana': 'Date','slippers: (Regne Unit)':'slippers'})

raw_boots['Date'] =  pd.to_datetime(raw_boots['Date'])
raw_sandals['Date'] =  pd.to_datetime(raw_sandals['Date'])
raw_goretex['Date'] =  pd.to_datetime(raw_goretex['Date'])
raw_slippers['Date'] =  pd.to_datetime(raw_slippers['Date'])

data['Date'] =  pd.to_datetime(data['Date'])
```

```python
data_merged = data.merge(raw_boots, how='left', on=['Date','style'])
data_merged = data_merged.merge(raw_sandals, how='left', on=['Date','style'])
data_merged = data_merged.merge(raw_slippers, how='left', on=['Date','style'])
data_merged = data_merged.merge(raw_goretex, how='left', on=['Date','style'])

#fill the nan
data_merged['boots'] = data_merged['boots'].fillna(0)
data_merged['sandals'] = data_merged['sandals'].fillna(0)
data_merged['slippers'] = data_merged['slippers'].fillna(0)
data_merged['goretex'] = data_merged['goretex'].fillna(0)
data_merged
```
